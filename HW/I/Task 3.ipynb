<p style="text-align: left">
    <h3>1</h3>
    Преобразуем уравнение $a y^3 + d = 0$ к следующему виду:
</p>
<p style="text-align: center">
    \[
        y = \sqrt[3]{{-d}\over{a}} = { \left( \frac{-d}{a} \right) }^{\frac{1}{3}};
    \]
</p>
<p style="text-align: left">
    Тогда выразим погрешность данного выражения следующим видом:
</p>
<p style="text-align: center">
    \[
        \begin{align*}
            \Delta(y) 
            =  \Delta { \left( \frac{-d}{a} \right) }^{\frac{1}{3}}
            = {\frac{1}{3}} \Delta \left( {\frac{-d}{a}} \right)
            = {\frac{1}{3}} {\frac{d}{a}} \left( \delta(a) + \delta(d) \right)
            = {\frac{d}{3a}} \left( \frac{\Delta(a)}{a} + \frac{\Delta(d)}{d} \right)
        \end{align*}
    \]
</p>
<p style="text-align: left">
    Вычислим погрешность
</p>

```python
import math
a = 1
d = 8

da = 1e-3
dd = 1e-3

y = math.cbrt(-d/a)
dy = d / (3*a) * (da/a + dd/d)

print(u"y \u00B1 \u0394y =", y, u"\u00B1", dy) 
```

<p style="text-align: left">
    <h3>2</h3>
    Знаем, что ошибка при счёте складывается из двух частей: ошибки метода $\varepsilon_{method}$, зависящей от выбранного, метода и ошибки связанной с вычислением значений функции $\varepsilon_{func}$:
</p>
<p style="text-align: center">
    \[
        \varepsilon = \varepsilon_{method} + \varepsilon_{func}.
    \]
</p>
<p style="text-align: left">
    Определим ошибки по отдельности, начнём с ошибки метода. Для этого поспользуемся разложением функции $u(x)$ в ряд Тейлора:
</p>
<p style="text-align: center">
    \[
        u(x+a) = u(x) + a \cdot u^{(1)}(x) + \frac{a^2}{2!} \cdot u^{(2)}(x) + \frac{a^3}{3!} \cdot u^{(3)}(x) +
        \frac{a^4}{4!} \cdot u^{(4)}(x) + \frac{a^5}{5!} \cdot u^{(5)}(x) + o(a^5),
    \]
</p>
<p style="text-align: left">
    тогда для нашей задачи:
</p>
<p style="text-align: center">
    \begin{align*}
        u(x-2h) &= u(x) &- 2h \cdot u^{(1)}(x) &+ \frac{4h^2}{2!} \cdot u^{(2)}(x) &- \frac{8h^3}{3!} \cdot u^{(3)}(x) &+
        \frac{16h^4}{4!} \cdot u^{(4)}(x) &- \frac{32h^5}{5!} \cdot u^{(5)}(x) &+ o(a^5);
        \\
        -8u(x-h) &= -8u(x) &+ 8h \cdot u^{(1)}(x) &- 8\frac{h^2}{2!} \cdot u^{(2)}(x) &+ 8\frac{h^3}{3!} \cdot u^{(3)}(x) &-
        8\frac{h^4}{4!} \cdot u^{(4)}(x) &+ 8\frac{h^5}{5!} \cdot u^{(5)}(x) &+ o(a^5);
        \\
        8u(x+h) &= 8u(x) &+ 8h \cdot u^{(1)}(x) &+ 8\frac{h^2}{2!} \cdot u^{(2)}(x) &+ 8\frac{h^3}{3!} \cdot u^{(3)}(x) &+
        8\frac{h^4}{4!} \cdot u^{(4)}(x) &+ 8\frac{h^5}{5!} \cdot u^{(5)}(x) &+ o(a^5);
        \\
        -u(x+2h) &= -u(x) &- 2h \cdot u^{(1)}(x) &- \frac{4h^2}{2!} \cdot u^{(2)}(x) &- \frac{8h^3}{3!} \cdot u^{(3)}(x) &-
        \frac{16h^4}{4!} \cdot u^{(4)}(x) &- \frac{32h^5}{5!} \cdot u^{(5)}(x) &+ o(a^5).
    \end{align*}
</p>
<p style="text-align: left">
    Рассмотрим почленное сложение всех этих выражений:
</p>
<p style="text-align: center">
    \[
        {u'(x)}_{method}:
    \]
</p>
<p style="text-align: center">
    \begin{align*}
        & При \space u(x): 0;
        \\
        & При \space u^{(1)}(x): \frac{12h}{12h} = 1 \space (т.е. \space искомая \space производная);
        \\
        & При \space u^{(2)}(x): 0;
        \\
        & При \space u^{(3)}(x): 0;
        \\
        & При \space u^{(4)}(x): 0;
        \\
        & При \space u^{(5)}(x): \frac{-48h^5}{12h \cdot 5!} = \frac{h^4}{30};
    \end{align*}
</p>
<p style="text-align: center">
    \begin{gather*}
        \Downarrow
        \\
        \varepsilon_{method} = \frac{h^4}{30} \cdot u^{(5)}(x) + o(u^{(5)}(x)) \leqslant \frac{h^4}{30} \cdot M_5.
    \end{gather*}
<p style="text-align: left">
    То есть наш метод даёт четвёртый порядок аппроксимации.
</p>
<p style="text-align: left">
    Теперь рассмотрим ошибку, связанную с точностью вычисления значений функции $\Delta(u)$. Схема численного дифференциирования четырёхточечная, ошибки вычислений совершаются независимо друг от друга, так что ошибку можно оценить через сумму ошибок:
</p>
<p style="text-align: left">
    \begin{multline*}
        {u'(x)}_{method}: \leqslant \Delta \biggl( \frac{u(x-2 h)-8 u(x-h)+8 u(x+h)-u(x+2 h)}{12 h} \biggr) = \\
        = \frac{\Delta u(x-2h) + 8 \Delta u(x-h) + 8 \Delta u(x+h) + \Delta u(x+2h)}{12h} =
        \frac{18 \Delta u }{12h} = \frac{3 \Delta u }{2h}.
    \end{multline*}
</p>
<p style="text-align: left">
    Итого, погрешность данной схемы выражается через сумму вышеописанных погрешностей. Заметим, что погрешность метода уменьшается с $h \rightarrow 0$, а другая погрешность уменьшается с ростом $h \rightarrow \infty$. Если инвертировать пределы, то погрешности неограниченно растут, следовательно, $\varepsilon(h)$ имеет минимум, найдём его:
</p>
<p style="text-align: left">
    \begin{gather*}
        \varepsilon'(h) = \Biggl( \frac{h^4}{30} \cdot M_5 + \frac{3 \Delta u }{2h} \Biggr)' = \frac{2h^3}{15} \cdot M_5 - \frac{3 \Delta u}{2h^2} = 0;
        \\
        h^5 = \frac{45 \Delta u}{4 M_5}.
        \\
        h_{opt} = \sqrt[5]{ \frac{45 \Delta u}{4 M_5} }
    \end{gather*}
</p>
