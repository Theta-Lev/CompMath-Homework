```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Ex 2.1
x = np.linspace(0, 10, 100)
y1 = x ** 2
y2 = np.sin(x)
y3 = np.exp(x / 3)
y4 = y1 + y2 * y3

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, y1, 'r')
axs[0, 0].set_title('y=x^2')

axs[0, 1].plot(x, y2, color='g', marker='o')
axs[0, 1].set_title('sin(x)')

axs[1, 0].plot(x, y3, 'b*', alpha=0.5)
axs[1, 0].set_title('exp(x/3)')

axs[1, 1].plot(x, y4, 'y--')
axs[1, 1].set_title('y = x^2 + sin(x)*exp(x /3)')

plt.tight_layout()

plt.show()
```

```python
#Ex 2.2

x = np.linspace(0, 10, 10)
y = np.sin(x)
dx = np.abs(np.atan(x)/np.cosh(x))
dy = np.abs(y*(np.sin(x))**2)

plt.errorbar(x, y, xerr = dx, yerr = dy, fmt='o', color='black', capsize=5)

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Title of the Graph')
plt.grid(True)

plt.show()
```

```python
#Ex 2.3

x = np.linspace(0, 10, 1000)
y = np.sin(np.exp(x))

plt.plot(x, y, 'g')
plt.xscale('log')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Title of the Graph')
plt.grid(True)

plt.show()
```
