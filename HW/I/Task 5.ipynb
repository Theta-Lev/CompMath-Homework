<ol>
    <h1>Упражнение 5 - Рост погрешности в последовательности</h1>
    <p>
        Пусть задана последовательность чисел $\{ x_n \}$, причем $5 x_{n+1} - x_n = 4$, а $x_0$ известно с относительной погрешностью $\delta x = 10^{-6}$. При каких значениях $x_0$ относительная погрешность при вычислении $x_n$ будет быстро возрастать с ростом $n$?
    </p>
</ol>
<hr>

<hr>
<p style="text-align: left">
    Преобразуем рекурентное выражение таким образом, чтобы выразить $(n+1)$-ый элемент последовательности через $n$-ый, затем сведём $x_{n+1}$ через $x_0$ и $n$:
</p>
<p style="text-align: center">
    \begin{align*}
        & 5 x_{n+1} - x_n = 4;
        \\
        & 5 x_{n+1} = 4 + x_n;
        \\
        & x_{n+1} = \frac{4}{5} + \frac{1}{5} \cdot x_n;
        \\
        & x_{n+1} = \frac{4}{5} + \frac{1}{5} \cdot \biggl( \frac{4}{5} + \frac{1}{5} \cdot x_{n-1} \biggr);
        \\
        & x_{n+1} = \frac{4}{5} \biggl( 1 + \frac{1}{5} \biggr) + {\biggl( \frac{1}{5} \biggr)}^{2} \cdot x_{n-1};
        \\
        & \vdots
        \\
        & x_{n+1} = \frac{4}{5} \sum_{k=0}^{n} {\left( \frac{1}{5} \right)}^{k} + {\left( \frac{1}{5} \right)}^{n+1} \cdot x_0.
    \end{align*}
</p>
<p style="text-align: left">
    Выразим абсолютную погрешность $x_{n}$ через абсолютную погрешность $x_0$, затем выразим относительную погрешность так, чтобы получить её зависимость от относительной погрешности $\delta x_0$ и номера $n$:
</p>
<p style="text-align: center">
    \begin{align*}
        & \Delta x_{n+1} = \Bigg| {\left( \frac{1}{5} \right)}^{n+1} \Bigg| \cdot \Delta x_0;
        \\
        & \delta x_{n+1} = \Bigg| \frac{ \Delta x_{n+1} }{ x_{n+1} } \Bigg| = 
        \Bigg| \frac{ {\left( \frac{1}{5} \right)}^{n+1} \cdot \Delta x_0 }{ \frac{4}{5} \sum_{k=0}^{n} {\left( \frac{1}{5} \right)}^{k} + {\left( \frac{1}{5} \right)}^{n+1} \cdot x_0} \Bigg| = 
        \Bigg| \frac{ {\left( \frac{1}{5} \right)}^{n+1} }{ \frac{4}{5 \cdot x_0} \sum_{k=0}^{n} {\left( \frac{1}{5} \right)}^{k} + {\left( \frac{1}{5} \right)}^{n+1} } \Bigg| \cdot \delta x_0
    \end{align*}
</p>
<p style="text-align: left">
    Заметим, что с ростом n сумма в знаменателе сходится к $\frac{5}{4}$, следовательно сам знаменатель сходится к 1/$x_0$. Для любого фиксированного $x_0 \in \mathbb{R}$ заметно, что числитель очень быстро сходится к нулю, следовательно невозможно подобрать такой $x_0$, что коэффициент перед $\delta x_0$ окажется больше единицы (т.е. относительная ошибка в определении $x_{n+1}$ будет больше, чем относительная ошибка $\delta x_0 = 10^{-6}$). Пробуя подставить в качестве значения $x_0 = \pm \infty$, получим, что коэффициент перед суммой в знаменателе равен $0$, что означает, что $\delta x_{n+1} = \delta x_0$. При попытке поставить $x_0 = 0$ в итоговую формулу получим неопределённость, разрешим её по более ранним формулам: 
</p>
<p style="text-align: center">
    \begin{align*}
        & x_{n+1} = \frac{4}{5} \sum_{k=0}^{n} {\left( \frac{1}{5} \right)}^{k} + {\left( \frac{1}{5} \right)}^{n+1} \cdot x_0 {\Bigg|}_{x_0 = 0} = \frac{4}{5} \sum_{k=0}^{n} {\left( \frac{1}{5} \right)}^{k} \rightarrow_{n \rightarrow \infty} 1.
    \end{align*}
</p>
<p style="text-align: left">
    При этом отметим, что при $x_0 = 0$ абсолютная погрешность $\Delta x_0 = 0$ и наше решение оказывается "абсолютно точным" по условию, то есть ни о каких абсолютных погрешностях вычислений членов последовательности речи идти не может. Таким образом, рассмотрены все случаи $x_0 \in \mathbb{R}$ и особые случаи $\pm \infty$.
</p>
<p style="text-align: left">
    <b>Итого</b>, ни при каких значениях $x_0$ относительная погрешность при вычислении членов последовательности $x_n$ не будет возрастать с ростом $n$.
</p>
